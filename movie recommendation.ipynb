```python
import numpy as np
import pandas as pd
import json
import ast
import pickle
```

```python
file1 = "E:/Movie recommendation/Movies Dataset/tmdb_5000_movies.csv"
file2 = "E:/Movie recommendation/Movies Dataset/tmdb_5000_credits.csv"
```

```python
movies = pd.read_csv(file1)
movies.head(1)
```

```python
credits = pd.read_csv(file2)
credits.head()
```

```python
movies = movies.merge(credits, on = "title")
movies.head()
```

```python
movies.info()
```

```python
df = pd.DataFrame(movies[['movie_id', 'title','genres', 'keywords', 'overview', 'release_date', 'status', 'cast', 'crew']])
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
def genresplit(genres):
    genre_list = json.loads(genres)
    return [genre["name"] for genre in genre_list]
```

```python
df.loc[:,'genres'] = df['genres'].apply(genresplit)
df.loc[:,'keywords'] = df['keywords'].apply(genresplit)
df.loc[:,'cast'] = df['cast'].apply(genresplit)
#df.loc[:,'crew'] = df['crew'].apply(genresplit)
df
```

```python
df.info()
```

```python
df['actors'] = df['cast'].apply(lambda x: x[:3])
df = df.drop(columns = ['cast'], axis = 0)
df
```

```python
def convertD(names):
    lst = []
    for i in ast.literal_eval(names):
        if i['job']== 'Director':
            lst.append(i['name'])
            break
    return lst
```

```python
df['crew'] = df['crew'].apply(convertD)
df['director'] = df['crew']
df = df.drop(columns = ['crew'], axis = 0)
df
```

```python
df['overview'] = df['overview'].str.split()
df['release_date'] = df['release_date'].str.split()
df['status'] = df['status'].str.split()
df
```

```python
def removeSp(L):
    L1 = [i.replace(" ", "") for i in L]
    return L1
```

```python
df.loc[:,'genres'] = df['genres'].apply(removeSp)
df.loc[:,'keywords'] = df['keywords'].apply(removeSp)
df.loc[:,'actors'] = df['actors'].apply(removeSp)
df.loc[:,'director'] = df['director'].apply(removeSp)
```

```python
df['data'] = df['overview'] + df['genres'] + df['keywords'] + df['release_date'] + df['status'] + df['actors'] + df['director']
df = df.drop(columns = ['genres','keywords','overview','release_date','status','actors','director'], axis = 0)
df
```

```python
df['data'] = df['data'].str.join(" ")
df
```

```python
df['data'][0]
```

```python
df['data'] = df['data'].str.lower()
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['data']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = similarity[index]
    movies_list = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])[1:6]
    for i in movies_list:
        print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
pickle.dump(df,open('movies.pkl','wb'))
pickle.dump(similarity,open('similarity_movies.pkl','wb'))
```
